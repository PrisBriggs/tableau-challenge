```python
# Dependencies
import pandas as pd
```

### Year 2019

```python
# Files to Load for year 2019
june_2019_data_to_load = "Resources/JC-201906-citibike-tripdata.csv"
july_2019_data_to_load = "Resources/JC-201907-citibike-tripdata.csv"
august_2019_data_to_load = "Resources/JC-201908-citibike-tripdata.csv"
september_2019_data_to_load = "Resources/JC-201909-citibike-tripdata.csv"
october_2019_data_to_load = "Resources/JC-201910-citibike-tripdata.csv"
november_2019_data_to_load = "Resources/JC-201911-citibike-tripdata.csv"

# Reading each month Data File and storing into Pandas DataFrames
june2019_data = pd.read_csv(june_2019_data_to_load)
july2019_data = pd.read_csv(july_2019_data_to_load)
august2019_data = pd.read_csv(august_2019_data_to_load)
september2019_data = pd.read_csv(september_2019_data_to_load)
october2019_data = pd.read_csv(october_2019_data_to_load)
november2019_data = pd.read_csv(november_2019_data_to_load)
```

```python
# Combining the data into a single dataset. 
y2019_df = pd.concat([june2019_data, july2019_data, august2019_data, 
                      september2019_data, october2019_data, november2019_data])
```

```python
# Cleaning up data 
# Making column titles the same for the four years chosen

y2019_renamed = y2019_df.rename(columns={"tripduration":"trip_duration_(s)",
                         "starttime": "started_at", 
                         "stoptime": "ended_at", 
                         "start station id": "start_station_id", 
                         "start station name": "start_station_name",
                         "start station latitude": "start_lat",
                         "start station longitude": "start_lng",
                         "end station id": "end_station_id",
                         "end station name": "end_station_name",
                         "end station latitude": "end_lat",
                         "end station longitude": "end_lng",    
                         "bikeid": "bike_id",
                         "usertype":"user_type",
                         "birth year": "birth_year"})
```

```python
# Replacing some values to more meaningful ones
y2019_renamed.loc[y2019_renamed['user_type'] == 'Subscriber', 'user_type'] = 'member'
y2019_renamed.loc[y2019_renamed['user_type'] == 'Customer', 'user_type'] = 'casual'
y2019_renamed.loc[y2019_renamed['gender'] == 0, 'gender'] = 'unknown'
y2019_renamed.loc[y2019_renamed['gender'] == 1, 'gender'] = 'M'
y2019_renamed.loc[y2019_renamed['gender'] == 2, 'gender'] = 'F'

y2019_renamed.head()
```

```python
# Writing dataset to a csv file. 
y2019_renamed.to_csv('year2019_jun_to_nov.csv', index=False)
```
